# Tweets scrapper

```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import json
```

## 1 Collect the tweets on 2021-02-26 with each of the term in the 1,300 words lexicon as the searching keyword

```python
lexicon_path = '../WebSci2019/lexicon.json'
```

```python
# Read the Farrell's keyword file
with open(lexicon_path) as f:
    lexicon = json.load(f)
```

```python
tweets_list = []
for category in lexicon:
    for keyword in lexicon[category]:
        print(category, keyword)
        for tweet in sntwitter.TwitterSearchScraper('{} since:2021-02-26 until:2021-02-27 lang:en'.format(keyword)).get_items():
            tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.user.location, tweet.place, tweet.coordinates, category, keyword]) #declare the attributes to be returned
```

```python
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'User', 'User_location', "Place", 'Coordinates', 'Category', 'Keyword'])
```

```python
tweets_list = []
for keyword in lexicon["Sexual Violence"]:
    print(keyword)
    for tweet in sntwitter.TwitterSearchScraper('{} since:2021-02-26 until:2021-02-27 lang:en'.format(keyword)).get_items():
        tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.user.location, tweet.place, tweet.coordinates, category, keyword]) #declare the attributes to be returned
```

```python
tweets_df_sexual_violence = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'User', 'User_location', "Place", 'Coordinates', 'Category', 'Keyword'])
```

```python
tweets_df_stoicism = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'User', 'User_location', "Place", 'Coordinates', 'Category', 'Keyword'])
```
